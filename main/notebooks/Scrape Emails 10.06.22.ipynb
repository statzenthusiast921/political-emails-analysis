```python
url_base = 'https://politicalemails.org/messages/'

email_pages=100000
for i in range(0,email_pages):

    url_full = url_base+str(i)
    print(f'This is iteration #: {i}')
    page = requests.get(url_full)
    if page.status_code == 404:
        continue

    soup = BeautifulSoup(page.text,'lxml')
    email = soup.find_all('td',class_='content-box-meta__value')
    extra_info = soup.find_all('ul',class_='key-val-list')

    message = soup.find_all('div',class_='message-text')
    if message == []:
        continue

    sender = email[0].text.strip()
    subject = email[1].text.strip()
    date = email[2].text.strip()

    body = message[0].text.strip()
    
    party = extra_info[0].findAll('li')[1].text.strip()
    country = extra_info[0].findAll('li')[2].text.strip()
    locality = extra_info[0].findAll('li')[3].text.strip()
    office = extra_info[0].findAll('li')[4].text.strip()

    df = df.append({
        'sender':sender, 
        'subject':subject,
        'date':date,
        'body':body,
        'party':party,
        'country':country,
        'locality':locality,
        'office':office
                
        },ignore_index=True)

df.shape
```

```python
df.to_csv('emails_df.csv')
```
